# Exercice 06 — Détection automatique d'intention par few-shot

**Module concerné :** Jour 2 — Module 6 (Few-shot & CoT)
**Durée estimée :** 25 minutes

## Contexte
Avant même de classer un ticket, tu veux détecter l'**intention** de l'utilisateur : est-ce une **demande d'information**, une **demande d'action** (créer/modifier quelque chose), ou un **signalement de panne** ?

## Consigne

1. Construire un `FewShotChatMessagePromptTemplate` avec **5 exemples diversifiés** (voir liste ci-dessous).
2. Le LLM doit retourner UN MOT parmi : `"information"`, `"action"`, `"panne"`.
3. Tester sur 5 nouvelles phrases (différentes des exemples) et afficher la prédiction.

## Exemples à utiliser
```python
exemples = [
    {"phrase": "Quelle est la procédure pour réinitialiser mon mot de passe ?", "intention": "information"},
    {"phrase": "Mon imprimante du 3e étage est en panne depuis ce matin.", "intention": "panne"},
    {"phrase": "Pourriez-vous créer un compte pour mon nouveau collaborateur Jean Martin ?", "intention": "action"},
    {"phrase": "Comment fonctionne le MFA chez nous ?", "intention": "information"},
    {"phrase": "Je voudrais demander un nouvel écran 27 pouces pour mon bureau.", "intention": "action"},
]
```

## Phrases de test
```python
phrases_test = [
    "Le serveur de fichiers est inaccessible depuis 30 minutes.",     # attendu : panne
    "Quels sont les horaires d'ouverture du support ?",               # attendu : information
    "Pouvez-vous m'ajouter à la liste de diffusion 'projets-2026' ?", # attendu : action
    "L'écran de Marie clignote toutes les 5 minutes.",                # attendu : panne
    "Je cherche la doc sur la politique de sauvegarde des PC.",       # attendu : information
]
```

## Critères de réussite
- Au moins **4 prédictions sur 5** sont correctes.
- Le code utilise bien `FewShotChatMessagePromptTemplate` (pas une string monolithique).

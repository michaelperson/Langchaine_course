**Module concerné :** Jour 2 — Module 7 (Agents & Tools)

## Contexte
Tu veux qu'un assistant puisse répondre à des questions du type :
*"Si je crée un ticket P2 maintenant, à quelle heure dois-je avoir traité ?"*

Pour ça, tu vas lui donner un **tool** qui calcule l'échéance SLA.

## Consigne

1. Définir 3 tools avec `@tool` :
   - `obtenir_heure_actuelle()` : retourne la datetime actuelle.
   - `calculer_echeance_sla(priorite: str)` : retourne la datetime d'échéance selon la grille SLA.
   - `lister_tickets_en_cours(categorie: str)` : retourne une liste fictive de tickets (simulation).
2. Définir la grille SLA dans la docstring du tool :
   - P1 : 1 heure
   - P2 : 4 heures
   - P3 : 1 jour ouvré
   - P4 : 3 jours ouvrés
3. Construire un `AgentExecutor` avec `create_tool_calling_agent`.
4. Tester sur 3 questions (voir ci-dessous).

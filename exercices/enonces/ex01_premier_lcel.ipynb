# Exercice 01 — Première chaîne LCEL

**Module concerné :** Jour 1 — Module 1 (LCEL)
**Durée estimée :** 15 minutes

## Contexte
Tu fais partie de l'équipe support de niveau 1. Ton manager te demande de créer un mini-outil interne qui prend en entrée la **description libre** d'un problème utilisateur et le **reformule** dans un style professionnel et concis (max 2 phrases) qui pourra être utilisé comme titre de ticket.

## Consigne

Construire une chaîne LCEL composée de :

1. Un `ChatPromptTemplate` qui :
   - donne au LLM le rôle de "rédacteur de tickets support",
   - lui demande de reformuler en max 2 phrases,
   - prend une variable `{description}`.
2. Le modèle `gemma4` via Ollama (URL via `os.environ["OLLAMA_HOST"]`).
3. Un `StrOutputParser`.

Ensuite, tester la chaîne sur les 3 descriptions ci-dessous et afficher le résultat.

```python
descriptions = [

```

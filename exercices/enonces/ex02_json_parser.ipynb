# Exercice 02 — Extracteur d'informations en JSON

**Module concerné :** Jour 1 — Module 2 (Prompts & Parsers)
**Durée estimée :** 20 minutes

## Contexte
L'équipe support reçoit des emails de signalement d'incidents. Tu veux extraire automatiquement les informations clés pour pré-remplir le formulaire de ticket.

## Consigne

Construire une chaîne LCEL qui prend un email brut et extrait un JSON avec les clés suivantes :

- `nom_demandeur` : le nom complet de la personne qui demande (str ou null si absent).
- `service` : le service mentionné (str ou null).
- `mots_cles` : liste des mots-clés techniques pertinents (list[str]).
- `sentiment` : `"calme"`, `"inquiet"`, `"frustre"` ou `"urgent"`.

Utilise :
- Un `ChatPromptTemplate` avec un system prompt clair.
- Un `JsonOutputParser` (n'oublie pas `format_instructions`).

## Données de test
```python
email = '''
Bonjour,

Je suis Marie Dupont, du service Comptabilité.

Depuis ce matin, je n'arrive plus à ouvrir Excel — il plante systématiquement quand
je clique sur le fichier "Bilan_Q1_2026.xlsx". J'ai redémarré mon poste, ça n'a rien
changé. C'est vraiment problématique car je dois finaliser le reporting trimestriel
pour vendredi !

Pourriez-vous m'aider rapidement svp ?

Marie
'''
```

## Critères de réussite
- La sortie est un **dict Python** (pas une str).
- Les 4 clés sont présentes.
- Le sentiment détecté est cohérent (probablement `"frustre"` ou `"urgent"`).
- Les mots-clés contiennent au moins `"Excel"` et le nom du fichier.

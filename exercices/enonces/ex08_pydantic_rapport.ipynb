# Exercice 08 — Validation Pydantic d'un rapport d'incident

**Module concerné :** Jour 2 — Module 8 (Pydantic & validation)
**Durée estimée :** 25 minutes

## Contexte
À chaque incident majeur (P1), un **rapport d'incident** structuré doit être produit. Tu veux qu'un LLM le génère automatiquement à partir d'une description libre, avec **validation stricte**.

## Consigne

1. Définir un modèle Pydantic `RapportIncident` avec :
   - `titre: str` (max 100 caractères).
   - `severite: Literal["P1", "P2"]`.
   - `services_impactes: list[str]` (au moins 1 élément).
   - `nb_utilisateurs_impactes: int` (>= 1).
   - `actions_immediates: list[str]` (3 à 5 actions).
   - `responsable: str` (l'équipe ou la personne).
   - `delai_resolution_estime_h: float` (entre 0.5 et 48).
2. Brancher ce modèle avec `llm.with_structured_output(RapportIncident)`.
3. Tester sur la description d'incident ci-dessous.

## Description de test
```python
description = '''
À 14h32 ce 12 mars 2026, le service Comptabilité a perdu l'accès au serveur ERP Sage X3.
30 personnes sont bloquées et ne peuvent ni saisir leurs factures ni générer leurs reportings.
La clôture mensuelle est prévue dans 3 jours. L'équipe Infrastructure investigue : ça ressemble
à une saturation disque sur le serveur applicatif. Pas d'impact sur les autres services.
Estimation initiale de remise en route : 2 à 4 heures.
'''
```

## Critères de réussite
- L'objet retourné est bien une instance de `RapportIncident`.
- Toutes les contraintes (Literal, ge, len) sont respectées (sinon ça lève une `ValidationError`).
- `services_impactes` contient au moins `"Comptabilité"` ou `"ERP Sage X3"`.

# Exercice 05 — Pipeline d'ingestion robuste

**Module concerné :** Jour 1 — Module 5 (Ingestion PDF/MD)
**Durée estimée :** 30 minutes

## Contexte
La DSI te demande d'industrialiser l'ingestion : tout fichier déposé dans `data/support_kb/` doit être indexé automatiquement, **sans doublon**, qu'il soit en PDF ou en Markdown.

## Consigne

1. Écrire une fonction `ingest(data_dir, persist_dir)` qui :
   - Charge tous les `.md` (avec `UnstructuredMarkdownLoader`).
   - Charge tous les `.pdf` (avec `PyPDFLoader`).
   - Combine les deux listes de Documents.
   - Découpe avec `RecursiveCharacterTextSplitter` (chunk_size=800, overlap=100).
   - Indexe dans Chroma avec un **id stable** par chunk (utilise `hashlib.md5(chunk.page_content.encode()).hexdigest()`).
   - Affiche un récap : `X docs / Y chunks indexés`.
2. Lancer la fonction sur `/workspace/data/support_kb` → `/workspace/data/chroma_db_exo5`.
3. Re-lancer la fonction **une 2e fois** sur les mêmes données — le compteur ne doit **PAS** doubler grâce aux IDs stables.

## Pourquoi les IDs stables ?
Sans ID, chaque exécution rajoute des doublons. Avec un ID basé sur le **hash du contenu**, Chroma fait un upsert au lieu d'un insert — c'est le pattern d'ingestion incrémentale en production.

## Critères de réussite
- 1ère exécution : N chunks indexés.
- 2e exécution : toujours N chunks (pas de doublon).
- Si tu modifies un fichier .md (ex: corriger une faute), seuls les chunks affectés ont un nouveau hash.

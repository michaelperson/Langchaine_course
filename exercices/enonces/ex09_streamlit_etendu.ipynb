# Exercice 09 — Étendre l'app Streamlit

**Module concerné :** Jour 2 — Module 9 (Streamlit)
**Durée estimée :** 30 minutes

## Contexte
L'application `examples/demo_streamlit_app.py` est fonctionnelle, mais ton manager te demande **3 améliorations** :

1. Ajouter un **bouton "Effacer la conversation"** dans la sidebar.
2. Afficher dans une zone dépliable (`st.expander`) **les sources** utilisées par le RAG pour chaque réponse.
3. Ajouter un **compteur** qui affiche le nombre de questions posées dans la session.

## Consigne

1. Crée une copie : `cp examples/demo_streamlit_app.py exercices/enonces/mon_app.py`.
2. Ouvre `mon_app.py` dans VSCode et implémente les 3 améliorations.
3. Lance ton app : `streamlit run exercices/enonces/mon_app.py --server.address=0.0.0.0`.

## Indications

### Pour le bouton "Effacer"
```python
if st.sidebar.button("Effacer la conversation"):
    st.session_state.messages = []
    st.rerun()
```

### Pour les sources
Modifie `build_rag_chain()` pour qu'elle renvoie un objet contenant **à la fois** la réponse et les docs sources. Ou bien, fais 2 appels distincts : un pour `retriever.invoke()`, un pour la chaîne de génération.

### Pour le compteur
Ajoute une clé dans `st.session_state` (ex: `nb_questions`) et incrémente-la à chaque nouvelle question.

## Critères de réussite
- Les 3 fonctionnalités marchent dans l'UI.
- La conversation est bien réinitialisée par le bouton.
- Les sources affichées sont **réellement** celles utilisées (pas un mock).
- Le compteur incrémente correctement.

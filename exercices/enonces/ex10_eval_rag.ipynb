# Exercice 10 — Eval automatique du RAG

**Module concerné :** Jour 2 — Module 10 (Debug & limites)
**Durée estimée :** 30 minutes

## Contexte
Avant de mettre l'assistant en prod, tu veux un **harnais de test automatique** qui mesure la qualité du RAG sur un jeu de questions de référence — **et qui peut tourner en CI**.

## Consigne

1. Créer un jeu d'eval avec 5-7 questions, chacune annotée avec :
   - `question`
   - `mots_cles_attendus` : liste de mots qui DOIVENT apparaître dans la réponse correcte.
   - `mots_cles_interdits` : liste de mots qui NE doivent PAS apparaître (signe d'hallucination).

2. Pour chaque question, calculer :
   - `recall = (nb mots-clés attendus trouvés) / (nb mots-clés attendus)` → entre 0 et 1.
   - `hallucination = 1 si un mot-clé interdit apparaît, sinon 0`.
   - `temps_execution_s` (avec `time.perf_counter()`).

3. Afficher un **tableau récapitulatif** propre (pandas DataFrame ou rich.Table).

4. Calculer un **score global** : moyenne des recalls et taux d'hallucination.
